## Variables

```python
x = 10
y = 3.5
name = "Manthan"
is_active = True

print("x : ",x)
print("y : ",y)
print("Name : ",name)
print(is_active)
```

## Numbers

```python
a = 10  #int
b = 3.14  #float
c = 123456789123456789 #long integer

print(a, " : " ,type(a))
print(b, " : " ,type(b))
print(c, " : " ,type(c))
```

## Arithmetic operations

```python
x = 10
y = 3

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x // y)  #floor division
print(x % y)   #remainder
print(x ** y)  #power
```
