## Reading Files

```python
f = open('input.txt','r')
print(f.read())
f.close()
```

```python
with open('input.txt','r') as f:
    print(f.read())
```

```python
f = open('input.txt','r')
```

```python
#read()
print(f.read())
```

```python
#read(n) - return 1st n characters
f.seek(0)
print(f.read(9))
```

```python
#readline() - returns one line

f.seek(0)  #moves file pointer to start of the file
print(f.readline())  #reads 1st line
print(f.readline())  #reads 2nd line
```

```python
#readlines() - returns list of all the sentences in file
f.seek(0)

lines = f.readlines()

for l in lines :
    print(l)
```

## Writing files

```python
# 'a' - Append - will append to the end of the file

with open('input.txt','a') as f:
    f.write('\nIhave updated this file!')
    f.close()
    
with open('input.txt','r') as f:
    print(f.read())
```

```python
# 'w' - Write - will overwrite any existing content

with open('input.txt','w') as f:
    f.write('\nI have overwritten this file!')  #write a single line
    f.close()
    
with open('input.txt','r') as f:
    print(f.read())
```

```python
# 'x' - will create a file, returns an error if the file exists

s = open('input.txt','x')
```

```python
# 'a','w' - will create a file if the specified file does not exists

s = open('output.txt','w')
s.write('This file is created by me.')

with open('output.txt','r') as s:
    print(s.read())
```

```python
#writelines() - write a list of strings to a file

with open('output.txt','w') as s:
    l = ['Hope you are doing well.','\nTake care!','\nByee!']
    s.writelines(l)
    s.close()
    
with open('output.txt', 'r') as s:
    print(s.read())
```

## Exception Handling

```python
for i in range(3):
    try:
        x = int(input("\nEnter number: "))
        y = 10 / x
    except ZeroDivisionError:
        print("Cannot divide by zero")
    except ValueError:
        print("Invalid input")
    else :
        print('Result : ',y)
```

```python
# finally - Always runs â€” whether an exception occurs or not.
try:
    f = open("input.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("File not found")
finally:
    print("Reading finished")
```

```python
#raise - Used to manually throw an exception.
age = int(input("Enter age: "))

if age < 18:
    raise ValueError("Age must be 18 or above")

print("You are eligible to vote")
```
