```python
student = {
    "name": "Manthan",
    "age": 21,
    "branch": "CSE"
}

print(student)
```

```python
# copy()
student_copy = student.copy()
print("Copy of dictionary:",student_copy)
```

```python
# get()
print("get('name'):", student.get("name"))
print("get('marks', 'Not Found'):", student.get("marks", "Not Found"))
```

```python
# keys()
print("keys():", student.keys())

# values()
print("\nvalues():", student.values())

# items()
print("\nitems():", student.items())
```

```python
# pop()
removed = student.pop("age")
print("After pop('age'):",student)
print("Removed value:", removed)
```

```python
# popitem()
print(student)
last = student.popitem()
print("After popitem():", student)
print("Removed pair:", last)
```

```python
# update()
print(student)
student.update({"marks": 90, "year": 3})
print("After update():", student)
```

```python
# fromkeys()
keys = ["id", "city", "country"]
new_dict = dict.fromkeys(keys, "Unknown")
print("fromkeys():", new_dict)
```

```python
# clear()
print(student)
student.clear()
print("\nAfter clear():")
print(student)
```
