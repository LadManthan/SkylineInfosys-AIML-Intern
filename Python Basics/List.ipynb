```python
nums = [10, 20, 30, 20, 40]
type(nums)
```

```python
#append - Adds one element at the end.
nums.append(50)
print(nums)
```

```python
#copy - creates a shallow copy
nums = [1, 2, 3]
new_nums = nums.copy()
print(new_nums)
```

```python
#clear - removes all elements
new_nums.clear()
print(new_nums)
```

```python
#count - Counts how many times a value appears.
nums = [10, 20, 30, 40, 50, 40]
print(nums.count(40))
```

```python
#extend - Adds multiple elements.
nums = [1, 2]
nums.extend([3, 4, 5])
print(nums)
```

```python
#insert - Insert at specific position.
nums = [10, 20, 30]
nums.insert(1, 40)
print(nums)
```

```python
nums = [10, 20, 30, 40]
```

```python
#index - Returns position of first occurrence
print(nums.index(20))
```

```python
#pop - Removes element by index.
nums.pop(1)
print(nums)
```

```python
#remove - Removes element by value.
nums.remove(30)
print(nums)
```

```python
#reverse - Reverses list order.
nums = [1,2,3,4,5]
nums.reverse()
print(nums)
```

```python
#sort - sorts the list
nums = [40, 10, 30, 20, 60, 50]
nums.sort()
print(nums)
```
