```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
def power(base, exp=2):
    return base ** exp

print(power(4))
print(power(4, 3))
```

```python
# variables passed to *args are stored as tuple
# variables passed to *kwargs are stored as dictionary

def details(*args, **kwargs):
    print(args)  #no. of arguments 
    print(type(args))
    
    print("\n")
    print(kwargs)  #no. of keyword arguments
    print(type(kwargs))

details(1, 2, 3, name="Manthan", age=21)
```

## Lambda Function

lambda arguments: expression

```python
square = lambda x: x * x
print(square(5))
```

```python
add = lambda a, b: a + b
print(add(3, 4))
```

### with built in functions

```python
nums = [1,2,3,4,5,6,7,8]
print(nums)
```

```python
#map() - Applies a function to each element of an iterable and returns the transformed results.
squares = list(map(lambda x : x*x, nums))
print(squares)
```

```python
#filter() - Selects elements from an iterable that satisfy a given condition.
even = list(filter(lambda x : x%2==0, nums))
print(even)
```

```python
#reduce() - Combines all elements of an iterable into a single value using a cumulative function.

from functools import reduce

prod = reduce(lambda a,b : a*b, nums)
print(prod)
```
