## Strings

```python
s = "This is a String"
print(type(s))
print(s)
```

```python
#Indexing

print(s[0]) #1st char
print(s[10]) #11th char
print(s[-1]) #last char
```

```python
#Slicing

print(s[0:7])
print(s[3:8])
print(s[:6])
print(s[5:])
print(s[::-1])   #reverse
```

```python
#updating
s = 'Archer Queen'
s = s.replace('Q','F')
print(s)

t = 'Tweet'
t = s.replace('e','x')
print(t)
```

```python
s = "this is random sentence."
```

```python
#Capitalize - converts 1st char to upper case
#Casefold - converts 1st char to lower case
print(s)

s = s.capitalize()
print(s)

s = s.casefold()
print(s)
```

```python
#count - Returns the number of times a specified value occurs in a string
print(s)
print(f"Count of e : {s.count('e')}")
```

```python
#find - searches the string for a specified value and returns the position of where it was found
idx = s.find('r')
print(idx)

#index - Searches the string for a specified value and returns the position of where it was found
x = s.index('r')
print(x)
```

```python
a = "alpha67"
print(f"Alpha numeric : {a.isalnum()}")  #Returns True if all characters in the string are alphanumeric

b = 'profit'
print(f"Alphabet : {b.isalpha()}")  #Returns True if all characters in the string are in the alphabet

c = '0A1'
print(f"Decimal : {c.isdecimal()}")  #Returns True if all characters in the string are decimal

d = '123456'
print(f"Numeric : {d.isnumeric()}")  #Returns True if all characters in the string are numeric
```

```python
#join - Joins the elements of an iterable to the end of the string

a = ("hello","this","is","a","msg")
print(" ".join(a))
print("_".join(a))
```

```python
b = "   cat   "
print(c,len(b))

c = b.strip()
print(c, len(c))

d = b.lstrip()  #Returns a left trim version of the string
print(d, len(d))

e = b.rstrip()  #Returns a right trim version of the string
print(e, len(e))
```

```python
s = """hello this is me.
how are?
hope you are fine."""

print(s.split(" "))  #Splits the string at the specified separator, and returns a list

print(s.splitlines())  #Splits the string at line breaks and returns a list
```

```python
#swapcase - Swaps cases, lower case becomes upper case and vice versa

f = "hEllO thIs Is A tExt"
print(f)
print(f.swapcase())
```

```python
#title - Converts the first character of each word to upper case

s = "how are you"
print(s.title())
```
