## Sets

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("A =", A)
print("B =", B)
```

```python
# add()
A.add(10)
print("\nAfter add(10) -> A:", A)
```

```python
# copy()
C = A.copy()
print("\nCopy of A -> C:", C)
```

```python
# clear()
temp = {7, 8, 9}
temp.clear()
print("\nAfter clear() -> temp:", temp)
```

```python
# difference()
print("\nA.difference(B):", A.difference(B))
print("A - B:", A - B)
```

```python
# difference_update()
D = A.copy()
D.difference_update(B)
print("\nAfter difference_update -> D:", D)
```

```python
# discard()
A.discard(10)
print("\nAfter discard(10) -> A:", A)
```

```python
# intersection()
print("\nA.intersection(B):", A.intersection(B))
print("A & B:", A & B)
```

```python
# intersection_update()
E = A.copy()
E.intersection_update(B)
print("\nAfter intersection_update -> E:", E)
```

```python
# isdisjoint()
print("\nA.isdisjoint({9,10}):", A.isdisjoint({9, 10}))
print("A.isdisjoint(B):", A.isdisjoint(B))
```

```python
# issubset()
X = {3, 4}
print("\nX <= A:", X <= A)
print("X < A:", X < A)
```

```python
# issuperset()
print("\nA >= X:", A >= X)
print("A > X:", A > X)
```

```python
# pop()
P = A.copy()
removed = P.pop()
print("\nAfter pop() -> removed:", removed, "Remaining:", P)
```

```python
# remove()
A.remove(4)
print("\nAfter remove(4) -> A:", A)
```

```python
# symmetric_difference()
print("\nA.symmetric_difference(B):", A.symmetric_difference(B))
print("A ^ B:", A ^ B)
```

```python
# symmetric_difference_update()
F = A.copy()
F.symmetric_difference_update(B)
print("\nAfter symmetric_difference_update -> F:", F)
```

```python
# union()
print("\nA.union(B):", A.union(B))
print("A | B:", A | B)
```

```python
# update()
G = A.copy()
G.update(B)
print("\nAfter update -> G:", G)
```
