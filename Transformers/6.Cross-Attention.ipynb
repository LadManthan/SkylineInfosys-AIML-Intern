# Cross Attetion
### Q is taken from output seq
### K,V are taken from output of Encoder block

```python
import numpy as np
```

```python
seq_len = 5
d_model = 6

x = np.random.rand(seq_len, d_model)
print(x)
print("Shape : ", x.shape)
```

## Create Q,K,V matrix for encoder output and Decoder input

```python
d_q = 3
d_kv = 4

#input to decoder
x_q = np.random.rand(d_q, d_model)
print("Decoder input : \n",x_q)
print("Shape : ", x_q.shape)

#output form encoder
x_kv = np.random.rand(d_kv, d_model)
print("\nEncoer Output : \n",x_kv)
print(x_kv.shape)
```

## Compute Q,K,V  matrix

```python
w_q = np.random.rand(d_model, d_model)
w_k = np.random.rand(d_model, d_model)
w_v = np.random.rand(d_model, d_model)
```

```python
Q = x_q @ w_q
K = x_kv @ w_k
V = x_kv @ w_v

print("Q : \n", Q)
print("Shape : ",Q.shape)

print("\nK : \n", K)
print("Shape : ",K.shape)

print("\nV : \n", V)
print("Shape : ",V.shape)
```

## Scaled Dot Product

```python
d_k = d_model 

score = Q @ K.T
score = score / d_k

print("Scaled dot product : \n",score)
print("Shape : ",score.shape)
```

## Softmax

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=1, keepdims=True))
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)
```

## Final Attention output

```python
attention_weights = softmax(score)
print("Attention weights : \n", attention_weights)
print("Shape : ",attention_weights.shape)
```

```python
attention_output = attention_weights @ V
print("Attention Output : \n",attention_output)
print("Shape : ",attention_output.shape)
```
