```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Line Plot

```python
roko = pd.read_csv('sharma-kohli.csv')
roko
```

```python
#linewidth - to set width of the line
#marker - to show the point on graph
#label - gives a name to the line
#color - to change color of line
#linestyle - to use different line styles

plt.plot(roko['index'],roko['V Kohli'],linewidth=2,marker='D',label='KO')
plt.plot(roko['index'],roko['RG Sharma'],linewidth=2, marker='o',label='RO')

#set title
plt.title('RO-KO Runs from 2008-2017')

#set x and y labels
plt.xlabel('Year')
plt.ylabel('Runs')

#to show legend
plt.legend()
```

# 3D Scatter Plot

```python
df = sns.load_dataset('tips')
df
```

```python
df['day'].unique()

df['day'] = df['day'].replace({'Thur':0, 'Fri':1, 'Sat':2, 'Sun':3})
df['time'] = df['time'].replace({'Lunch':0,'Dinner':1})
df.head()
```

```python
plt.scatter(df['total_bill'],df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')

ax.scatter3D(df['day'],df['total_bill'],df['tip'],color='green')

ax.set_xlabel('Day')
ax.set_ylabel('Total Bill')
ax.set_zlabel('Tip')
ax.set_title('Total Bill vs Tips on each Day')
```

# 3D Line Plot

```python
x = [2.2,0,10,8,2.2]
y = [0.8,1,2,0.2,0.8]
z = [0,2,3,0,0]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100,100],color='green')
ax.plot3D(x,y,z,color='red')
ax.set_title('3D Line Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

# 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)

z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

# Contour Plot

```python
fig = plt.figure(figsize=(10,8))
ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title('Contour Plot')
```

```python
fig = plt.figure(figsize=(10,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title('Contour Fill')
```

# Heat Map

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]

grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(18,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()

plt.xlabel('Ball No.')
plt.ylabel('Over No.')
plt.title('No. of sixes on each ball in an over')
```
