### .apply() - applies a function to dataframe

```python
points = pd.DataFrame(
    {
        '1st point':[(3,5),(-7,5),(0,0),(-12,1),(6,5)],
        '2nd point':[(-3,4),(0,4),(12,2),(10,3),(1,1)]
    }
)
points
```

```python
def distance(row):
    x = row['1st point']
    y = row['2nd point']
    
    return ((x[0] - y[0])**2 + (x[1] - y[1])**2)
```

```python
points['Distance'] = points.apply(distance, axis=1)
points
```
