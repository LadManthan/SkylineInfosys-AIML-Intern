## Feature Engineering

#### Creating new col - family size

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
df.head(3)
```

#### Adding categorical col - family size

```python
def family_type(num):
    if num == 1:
        return 'alone'
    elif num>1 and num <5:
        return "small"
    else:
        return "large"
```

```python
df['Family_type'] = df['Family_size'].apply(family_type)
df.head(3)
```
