# Tree

```python
class node:
    def __init__(self,val,left=None,right=None):
        self.val = val
        self.left = None
        self.right = None
        
    def Insert(self,val):
        new = node(val)
        
        if self is None:
            return new
        if val < self.val :
            if self.left is None:
                self.left = new
            else :
                self.left.Insert(val)
        else :
            if self.right is None:
                self.right = new
            else :
                self.right.Insert(val)
        return self
    
    def Preorder(self):
        if self is None:
            return
        
        print(self.val, end=" ")
        
        if self.left:
            self.left.Preorder()
        if self.right:
            self.right.Preorder()
            
    def Inorder(self):
        if self is None:
            return
        if self.left :
            self.left.Inorder()
        print(self.val, end=" ")
        
        if self.right :
            self.right.Inorder()
    
    def Postorder(self):
        if self is None :
            return
        if self.left :
            self.left.Postorder()
        if self.right :
            self.right.Postorder()
        print(self.val, end=" ")
        
    def DFS(self):
        stack = [self]
        dfs=[]
        while stack:
            x = stack.pop()
            dfs.append(x.val)
            
            if x.right :
                stack.append(x.right)
            if x.left :
                stack.append(x.left)
        return dfs
            
    def BFS(self):
        queue = [self]
        bfs=[]
        while queue:
            x = queue.pop(0)
            bfs.append(x.val)
            
            if x.left :
                queue.append(x.left)
            if x.right :
                queue.append(x.right)
        return bfs
```

```python
root = node(10)
a=[4,15,6,12,11,1,3,7,13]

for i in range(len(a)):
    root.Insert(a[i])

print(root)
```

### Preorder : Visit -  Left -  Right

```python
print("Preorder Traverssal :")
root.Preorder()
```

### Inorder : Left - Visit - Right

```python
print("Inorder Traverssal :")
root.Inorder()
```

### Postorder : Left - Right - Visit

```python
print("Postorder Traverssal :")
root.Postorder()
```

### Depth First

```python
print("DFS Traversal : ")
x = root.DFS()
print(x)
```

### Breadth First

```python
print("BFS Traversal : ")
y = root.BFS()
print(y)
```
