# singly Linked List 

```python
class node:
    def __init__(self,data):
        self.data = data
        self.next = None
    
    def insertfirst(self,data):
        new = node(data)
        new.next = self
        return new
    
    def insertend(self,data):
        new = node(data)
        if self is None:
            return new
        temp = self
        while temp.next:
            temp = temp.next
        temp.next = new
        return self
    
    def deletefirst(self):
        if self is None:
            print("LL EMPTY !!")
        self = self.next
        return self
        
    def deleteend(self):
        if self is None:
            print("LL EMPTY !!")
        temp = self
        while temp.next.next:
            temp=temp.next
        temp.next = None
        return self
    
    def search(self,x):
        self.display()
        count = 0
        temp = self
        while temp:
            if temp.data == x:
                print(f"{x} found at position {count+1}")
                return self
            else:
                count += 1
                temp = temp.next
        print("element not present in LL")
        return self
    
    def display(self):
        string=[]
        temp = self
        while temp:
            string.append(str(temp.data))
            temp = temp.next
        print(' --> '.join(string))
```

# creating node

```python
head = node(0)
head.display()
```

# inserting at front

```python
head = head.insertfirst(1)
head = head.insertfirst(2)
head = head.insertfirst(3)

head.display()
```

# inserting at end

```python
head = head.insertend(1)
head.display()
```

# deleting first node 

```python
head = head.deletefirst()
head.display()
```

# deleting end node 

```python
head = head.deleteend()
head.display()
```
