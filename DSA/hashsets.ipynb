# Hash Sets 

```python
s = set()
print(s)
```

```python
#adding item into set - O(1)
s.add(1)
s.add(2)
s.add(3)

print(s)
```

```python
#lookup in set - O(1)

if 1 in s:
    print(True)
```

```python
#removing item - O(1)
s.remove(3)
print(s)
```

```python
string = 'aaaaaaabbbbbbbbcccddddddddddeee'
sett = set(string)   #takes only unique values
print(sett)
```

# Hash Maps (Dictionaries)

```python
d = {'manthan' : 1,'j':2,'lad':3}
print(d)
```

```python
#adding key : value in dic - O(1)

d['engineer'] = 4
print(d)
```

```python
print(d['manthan'])
```

```python
#lookup key : valuepairs in dic - O(n)

for key,val in d.items() :
    print(f"{key} : {val}")
```

# Defaultdict

```python
from collections import defaultdict
```

```python
default = defaultdict(int)
default[3]   #maps 3 to 0 
```

```python
for key,val in default.items() :
    print(f"{key} : {val}")
```

# Counter

```python
from collections import Counter
```

```python
str = 'aaaaaaaabbbbccddddddddddddddddeeeeeee'

count = Counter(str)    #creturns count of each element
print(count)

for i in count:
    print(count[i])
```

```python
nums=[2,14,18,22,22]
num = Counter(nums)

i = int(input("Enter no. : "))
print(f"Count of {i} : {num[i]}")
```
